# SEABORN ASSIGNMENT

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic
dataset.

```python
#solution 1
df = pd.DataFrame({"x":(1,2,3,4,5,6,7,8,9),"y":(3,4,5,6,7,8,9,1,2)})
sns.scatterplot(x = df.x,y = df.y,data = df)

```

2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

```python
x = list(np.random.randn(50))
sns.histplot(x,kde = True)   
#clearly given distribution is normal distribution
```

3. Create a dataset representing categories and their corresponding values. Compare different categories
based on numerical values.

```python
#solution 3
df = pd.DataFrame((["A","B","C","D","E"],[67,78,87,83,76]),index = None)
df = df.T
df.columns = ["sections","avg_score"]
df
```

```python
sns.barplot(x = df.sections,y = df.avg_score,data = df,palette = "pastel")
```

4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical

```python
#solution 4
df = sns.load_dataset("tips")
df.head(2)
```

```python
sns.violinplot(data=df, x="time", y="tip",)
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
#solution 5
df= sns.load_dataset("iris")
df.head(2)
```

```python
df= df.drop('species',axis = 1)
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(),annot = True)
```
